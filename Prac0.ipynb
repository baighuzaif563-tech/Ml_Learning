```python

class Node:
    def __init__(self, task):
        self.task = task
        self.next = None


class ManageTodo:
    def __init__(self):
        self.head = None

    def addTask(self, task):
        newNode = Node(task)

        if self.head == None:
            self.head = newNode
        else:
            current = self.head
            while current.next != None:
                current = current.next
            current.next = newNode

    def displayTask(self):
        current = self.head

        if current == None:
            print("No Tasks")
        else:
            while current != None:
                print(current.task)
                current = current.next

    def removeTask(self, task):
        current = self.head
        previous = None

        while current != None:
            if current.task == task:
                if previous == None:
                    self.head = current.next
                else:
                    previous.next = current.next
                print("Task Removed")
                return

            previous = current
            current = current.next

        print("Task Not Found")


todo = ManageTodo()

while True:
    print("\n1. Add Task")
    print("2. Remove Task")
    print("3. Display Tasks")
    print("4. Exit")

    choice = int(input("Enter Choice: "))

    if choice == 1:
        task = input("Enter Task: ")
        todo.addTask(task)

    elif choice == 2:
        task = input("Enter Task to Remove: ")
        todo.removeTask(task)

    elif choice == 3:
        todo.displayTask()

    elif choice == 4:
        break

    else:
        print("Invalid Choice")










```

```python
def encrypt(text, shift):
    result = ""

    for ch in text:
        if ch.isalpha():
            result += chr(ord(ch) + shift)
        else:
            result += ch

    return result


def decrypt(text, shift):
    result = ""

    for ch in text:
        if ch.isalpha():
            result += chr(ord(ch) - shift)
        else:
            result += ch

    return result


choice = int(input("1. Encrypt\n2. Decrypt\nEnter Choice: "))
shift = 3

if choice == 1:
    file = open("input.txt", "r")
    data = file.read()
    file.close()

    encrypted = encrypt(data, shift)

    file = open("encrypted.txt", "w")
    file.write(encrypted)
    file.close()

    print("File Encrypted Successfully!")

elif choice == 2:
    file = open("encrypted.txt", "r")
    data = file.read()
    file.close()

    decrypted = decrypt(data, shift)

    file = open("decrypted.txt", "w")
    file.write(decrypted)
    file.close()

    print("File Decrypted Successfully!")

else:
    print("Invalid Choice")
```

```python
graph = {}

def addFriend(person, friend):
    if person not in graph:
        graph[person] = []

    if friend not in graph:
        graph[friend] = []

    graph[person].append(friend)
    graph[friend].append(person)


def displayFriends(person):
    if person in graph:
        print("Friends:", graph[person])
    else:
        print("Person not found")


def mutualFriends(p1, p2):
    if p1 in graph and p2 in graph:
        for friend in graph[p1]:
            if friend in graph[p2]:
                print(friend)


while True:
    print("\n1. Add Friend")
    print("2. Display Friends")
    print("3. Mutual Friends")
    print("4. Exit")

    choice = int(input("Enter Choice: "))

    if choice == 1:
        p1 = input("Enter Person: ")
        p2 = input("Enter Friend: ")
        addFriend(p1, p2)

    elif choice == 2:
        person = input("Enter Person: ")
        displayFriends(person)

    elif choice == 3:
        p1 = input("First Person: ")
        p2 = input("Second Person: ")
        mutualFriends(p1, p2)

    elif choice == 4:
        break

    else:
        print("Invalid Choice")
```
